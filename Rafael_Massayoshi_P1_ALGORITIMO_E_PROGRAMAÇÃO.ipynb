```python
n1 = int(input("Informe um numero: "))
n2 = int(input("Informe outro numero: "))

#soma
soma = n1 + n2
print ("A soma destes dois numeros é igual a:",soma)

#Subtração
sub = n1 - n2
print ("A subtração destes dois numeros é igual a:",sub)

#divisão
div = n1 / n2
print ("A divisão destes dois numeros é igual a:",div)

#Multiplicação
mult = n1 * n2
print ("A multiplicação destes dois numeros é igual a:",mult)
```

3- Um novo modelo de carro, super econômico foi lançado. ele faz 20km com 1 litro de combustivel. Cada litro de combustivel custa R$5,00. faça um programa que pergunte ao usuario quanto de dinheiro ele tem e em seguida diga quantos litros de combustivel ele pode comprar e quantos kilometros o carro consegue andar com esta quantidade de combustivel

```python
#dinheiro disponivel
dinheiro = int(input("informe quanto dinheiro voce tem: "))

kml = 20
preco = 5

#quantos litros o cliente pode comprar de combustivel
podecomprar = dinheiro / preco
print ("voce pode comprar",podecomprar,"litros")

#Quntos kilometros o carro consegue andar
andar = kml * podecomprar
print ("o carro podera andar",andar,"kilometros")
```
